```python
def solve_n_queens(n):
    def is_safe(board, row, col):
        # Controlla la riga
        for i in range(col):
            if board[row][i] == 1:
                return False
        
        # Controlla la diagonale superiore sinistra
        for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
            if board[i][j] == 1:
                return False
        
        # Controlla la diagonale inferiore sinistra
        for i, j in zip(range(row, n, 1), range(col, -1, -1)):
            if board[i][j] == 1:
                return False
        
        return True

    def solve(board, col):
        if col >= n:
            return True
        
        for i in range(n):
            if is_safe(board, i, col):
                board[i][col] = 1
                if solve(board, col + 1):
                    return True
                board[i][col] = 0
        
        return False

    board = [[0 for _ in range(n)] for _ in range(n)]
    
    if solve(board, 0):
        return board
    else:
        return None

# Esempio di utilizzo
solution = solve_n_queens(8)
if solution:
    for row in solution:
        print(" ".join("X" if x else "0" for x in row))
else:
    print("Nessuna soluzione trovata.")
```

```python
import numpy as np
from scipy.optimize import minimize

def objective(x):
    board = x.reshape(8, 8)
    
    # Penalizza valori non binari
    binary_penalty = np.sum((board * (1 - board))**2)
    
    # Penalizza più di una regina per riga/colonna
    row_penalty = np.sum((np.sum(board, axis=1) - 1)**2)
    col_penalty = np.sum((np.sum(board, axis=0) - 1)**2)
    
    # Penalizza regine sulla stessa diagonale
    diag_penalty = 0
    for i in range(-7, 8):
        diag_sum = np.sum(np.diag(board, k=i))
        diag_penalty += diag_sum * (diag_sum - 1)
    
    anti_diag_penalty = 0
    for i in range(-7, 8):
        anti_diag_sum = np.sum(np.diag(np.fliplr(board), k=i))
        anti_diag_penalty += anti_diag_sum * (anti_diag_sum - 1)
    
    return binary_penalty + row_penalty + col_penalty + diag_penalty + anti_diag_penalty

# Vincoli per mantenere i valori tra 0 e 1
bounds = [(0, 1) for _ in range(64)]

# Inizializzazione casuale
x0 = np.random.rand(64)

# Ottimizzazione
result = minimize(objective, x0, method='trust-constr', bounds=bounds)

# Arrotondamento del risultato
final_board = np.round(result.x.reshape(8, 8))

print("Soluzione trovata:")
print(final_board)
print("Valore della funzione obiettivo:", objective(result.x))

# Verifica se la soluzione è valida
if np.sum(final_board) == 8 and objective(final_board.flatten()) < 1e-6:
    print("Soluzione valida trovata!")
else:
    print("La soluzione non è completamente valida.")
```
